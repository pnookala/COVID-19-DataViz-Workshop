## Definitions

*  S = Susceptibles
*  I = Infected
*  R = Recovered (Removed)

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

#  Initialize Table for Model

```
number_of_time_periods = 24 * 3 * 30   # hourly for 3 months
time_range_hours = range(0, number_of_time_periods)
time_range_days = [ hour / 24.0 for hour in time_range_hours]

regional_population = 12000000.0   #  Close to the Illinois population

S = regional_population
I = 1.0
R = 0.0

initial_conditions = { 'Susceptibles' : [S] * number_of_time_periods, 'Infected' : [I]*number_of_time_periods, 'Recovered': [R]*number_of_time_periods}

sir = pd.DataFrame(data=initial_conditions, index=time_range_days)

sir.head(10)
```

#  Define Model Parameters

```
social_contact_per_day = 0.4
social_contact_per_hour = social_contact_per_day / 24.0

infection_rate_per_hour = social_contact_per_hour / regional_population

recovery_rate_per_day = 0.09
recovery_rate_per_hour = recovery_rate_per_day / 24.0

B = infection_rate_per_hour
L = recovery_rate_per_hour
```

#  Simplified calculation of R0

```
R0 = S * B / L
```

#  More on R0 calculations



"*The expected number of cases directly generated by one case in a population where all individuals are susceptible to infection*"

*  "Basic reproduction number"  https://en.wikipedia.org/wiki/Basic_reproduction_number
*  "Theory versus Data: How to Calculate R0?"    https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1804098/
*  "Complexity of the Basic Reproduction Number (R0)"  https://wwwnc.cdc.gov/eid/article/25/1/17-1901_article



#  Running the SIR Model

```
for t in time_range_hours: 

  sir.iloc[t]['Susceptibles'] = S
  sir.iloc[t]['Infected'] = I
  sir.iloc[t]['Recovered'] = R

  new_infections = B * I * S
  new_recoveries = L * I

  dS = - new_infections
  dI =   new_infections - new_recoveries
  dR =                    new_recoveries

  S = S + dS
  I = I + dI
  R = R + dR
```

#  Simple plot of the populations over time

```
sir.plot.line(title='SIR Model', figsize=[25,15]);
```

#  Exponential process are better seen in Logarithmic scales

```
sir.plot.line(title='SIR Model - Logarithmic Scale', figsize=[25,15], logy=True);
```

# Exponential Function and its Variation across Time

# $y(t) = e^{B x(t)}$

# $ \frac{dy(t)}{dt} = B x(t) e^{B x(t)} \frac{dx(t)}{dt}$

 # $ \frac{dy(t)}{dt} = B x(t) y(t) \frac{dx(t)}{dt}$

# For the early portion of the Kermack-McKendrick Model

 # $ \frac{dI(t)}{dt} = B S(t) I(t) \frac{dS(t)}{dt}$

#  Comparison with Actual Data in Illinois

```
cases_history = pd.read_csv("https://projects.rcc.uchicago.edu/rcc/covid19-data/covid-19-illinois-data/historical_state_testing_results.csv")
cases_history.set_index('testDate', drop=True, inplace=True)
confirmed_cases = cases_history['confirmed_cases']
confirmed_cases.plot.line(rot=45, subplots=True, figsize=[25,15]);
```

#  Selecting a subset of the simplified SIR model data to compare with real data

* THIS IS A **PURELY ACADEMIC EXERCISE**
* The simplified SIR model above is **NOT** based on any real data

```
time_from_beginning_of_record = dt.date.today() - dt.date(2020, 3, 10)
days_from_beginning_of_record = time_from_beginning_of_record.days

sir_mapped = [ sir['Infected'].iloc[day*24] for day in range(0,days_from_beginning_of_record)]
infected_cases = pd.Series(data=sir_mapped, name='infected', index=confirmed_cases.index)
ic = pd.concat([confirmed_cases, infected_cases], axis=1)
```

#  Plotting real data versus SIR textbook example

```
fig, axes = plt.subplots(1,2, figsize=(30, 8))
fig.suptitle('Exponential Growth', fontsize=20)

ic.plot.line(ax=axes[0], title='Linear Scale');
ic.plot.line(ax=axes[1], title='Logarithmic Scale', logy=True);
```

#  Computing the "time to double"

```
number_of_cases_log2 = np.log2(ic)      #  Log in base 2, to get   y = 2^x   

window_size = 5
smoothed_number_of_cases_log2 = number_of_cases_log2.rolling(window=window_size, win_type='boxcar').sum() / window_size

derivative_number_of_cases_log2 = smoothed_number_of_cases_log2.diff()

days_to_double = 1.0 / derivative_number_of_cases_log2      #    "Instantaneous speed"

fig, axes = plt.subplots(1,2, figsize=(30, 8))
fig.suptitle('Time to Grow by factor of 2', fontsize=20)

smoothed_number_of_cases_log2.plot.line(ax=axes[0], title='Slope of Log(number of cases)')
days_to_double.plot.line(ax=axes[1], title='Days to Double ');
```
